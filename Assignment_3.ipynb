```python
#Q4. checks if a given number is prime or not from 1 to 200.
def is_prime(n):
  if n <= 1:
    return False
  for i in range(2, int(n**0.5) + 1):
    if n % i == 0:
      return False
  return True

# Example usage:
for num in range(1, 201):
  if is_prime(num):
    print(num, end=" ")
```

```python
#Q5.Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms?

class FibonacciIterator:
    def __init__(self, num_terms):
        self.num_terms = num_terms
        self.current_term = 0
        self.a, self.b = 0, 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.current_term < self.num_terms:
            if self.current_term <= 1:
                result = self.current_term
            else:
                result = self.a + self.b
                self.a = self.b
                self.b = result
            self.current_term += 1
            return result
        else:
            raise StopIteration

# Example usage:
fib_iter = FibonacciIterator(10)
for num in fib_iter:
    print(num)
```

```python
#Q6.Write a generator function in Python that yields the powers of 2 up to a given exponent?
def powers_of_two(exponent):
  for i in range(exponent + 1):
    yield 2**i

# Example usage:
for power in powers_of_two(5):
  print(power)
```

```python
#Q7.Implement a generator function that reads a file line by line and yields each line as a string.

def read_file_line_by_line(file_path):
  """
  Reads a file line by line using a generator.

  Args:
    file_path: The path to the file.

  Yields:
    Each line of the file as a string.
  """
  with open(file_path, 'r') as f:
    for line in f:
      yield line.strip() # strip() removes leading/trailing whitespace, including the newline character

# Example usage:
file_name = "sample.txt"
for line in read_file_line_by_line(file_name):
  print(line)
```
